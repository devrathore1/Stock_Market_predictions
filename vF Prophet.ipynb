```python
import pandas as pd
import numpy as np
```

```python
#read the data
df = pd.read_excel(
     "/content/drive/MyDrive/823/DMMA823_AssignmentData_topost.xlsx")
df
```

```python
#select date & price
df = df[['Date', 'MMAX']]
df = df.rename(columns={'Date': 'ds', 'MMAX': 'y'})
```

```python
#Get some test data and store in new variables to check rmse later
last = df[len(df)-20:]
last
```

```python
#get test data
df = df[:-20]
df.tail(10)
```

```python
#Install prophet and prophet object
!pip install pystan~=2.14
!pip install fbprophet
from fbprophet import Prophet

fbp = Prophet(daily_seasonality= True)


```

```python
#Fit
fbp.fit(df)
future = fbp.make_future_dataframe(periods = 20)
forecast = fbp.predict(future)
```

```python
#import fb prophet plot libraries 
from fbprophet.plot import plot_plotly
#plot the data
plot_plotly(fbp, forecast)
```

```python
#Prediction value for 1973-01-25

forecast[forecast.ds == '1973-01-25']['yhat']
```
